```python
import matplotlib.pyplot as plt
import numpy as np
#so i can see in dark mode
plt.style.use(['dark_background'])
x = np.linspace(0, 2*np.pi, 1001)
y = np.sin(x)
plt.plot(x, y)
#so i can see in dark mode
plt.suptitle('figure title', color='w')
plt.savefig("image-example.png")
```
